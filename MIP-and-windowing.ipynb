```python
import numpy as np
import nibabel as nb
import matplotlib.pyplot as plt
```

```python
# code to check the working of thresholding

im = '51'
# 49, 51, 52, 81

og_data_ct = nb.load('ct_scans/0'+im+'.nii')
type(og_data_ct), og_data_ct.shape

data_ct = og_data_ct.get_fdata()

data_ct_th = np.multiply(data_ct, (data_ct>50)*1 & (data_ct<2000)*1)   # applying threshold

i = int(og_data_ct.shape[2]/2)
i = 25


a = (data_ct[:, :, i])
b = a.copy(); b = ((b - np.min(b))/(np.max(b) - np.min(b)))*255      # normalizing valus of the ct_scan from 0 to 255 for showing the image
print(a.shape, b.shape)

c = (data_ct_th[:, :, i])
d = c.copy(); d = ((d - np.min(d))/(np.max(d) - np.min(d)))*255      # normalizing valus of the ct_scan from 0 to 255 for showing the image
print(a.shape, b.shape)


plt.figure(figsize=(20, 20))
plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.xlabel('(a)', fontsize = 30)
plt.subplot(1, 3, 2)
plt.imshow(d, cmap='gray')
plt.xlabel('(b)', fontsize = 30)
plt.subplot(1, 3, 3)
data_ct_th = np.multiply(data_ct, (data_ct<300)*1)
e = (data_ct_th[:, :, i])
f = e.copy(); f = ((f - np.min(f))/(np.max(f) - np.min(f)))      # normalizing valus of the ct_scan from 0 to 255 for showing the image
plt.imshow(f, cmap='gray')
plt.xlabel('(c)', fontsize = 30)
```

```python
# code to generate MIP imags after thresholding

im = '81'
# 49, 51, 52, 81

og_data_ct = nb.load('ct_scans/0'+im+'.nii')
type(og_data_ct)
print(og_data_ct.shape)

data_ct = og_data_ct.get_fdata()

data_ct_th = np.multiply(data_ct, (data_ct>300)*1 & (data_ct<2000)*1)   # applying threshold

i = int(og_data_ct.shape[2]/2)

a = np.max(data_ct[:, :, i::], axis=2)
b = a.copy(); b = ((b - np.min(b))/(np.max(b) - np.min(b)))*255      # normalizing valus of the ct_scan from 0 to 255 for showing the image
print(a.shape, b.shape)

c = np.max(data_ct_th[:, :, i::], axis=2)
d = c.copy(); d = ((d - np.min(d))/(np.max(d) - np.min(d)))*255      # normalizing valus of the ct_scan from 0 to 255 for showing the image
print(a.shape, b.shape)

plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(b, cmap='gray')
plt.xlabel('(a)', fontsize = 30)
plt.subplot(1, 2, 2)
plt.imshow(d, cmap='gray')
plt.xlabel('(b)', fontsize = 30)
```
